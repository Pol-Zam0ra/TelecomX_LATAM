#ðŸ”§ TransformaciÃ³n

##**Manejo de inconsistencias**

```python
df['Churn'] = df['Churn'].str.strip().str.lower().replace('', 'yes')
df['Churn'].unique()
```

```python
df.head()
```

```python
df['customer_gender'] = df['customer'].apply(lambda x: x['gender'].strip().capitalize())
df['customer_gender'].unique()
```

```python
df['customer_senior'] = df['customer'].apply(lambda x: int(x['SeniorCitizen']))
df['customer_senior'].value_counts()
```

```python
df['charges_monthly'] = df['account'].apply(lambda x: x['Charges']['Monthly'] if 'Charges' in x and 'Monthly' in x['Charges'] else None)
df['charges_monthly'] = pd.to_numeric(df['charges_monthly'], errors='coerce')
df['charges_monthly'].head()
```

```python
df['charges_total'] = df['account'].apply(lambda x: x['Charges']['Total'] if 'Charges' in x and 'Total' in x['Charges'] else None)
df['charges_total'] = pd.to_numeric(df['charges_total'], errors='coerce')
df['charges_total'].head()
```

```python
df['contract'] = df['account'].apply(lambda x: x.get('Contract', None))
df['contract'].head()
```

```python
df['contract'] = df['contract'].str.strip().str.lower()
```

```python
df['payment_method'] = df['account'].apply(lambda x: x.get('PaymentMethod', None))
df['payment_method'].head()
```

```python
df['internet_service'] = df['internet'].apply(lambda x: x.get('InternetService', None))
df['internet_service'].head()
```

```python
df['tech_support'] = df['internet'].apply(lambda x: x.get('TechSupport', None))
df['tech_support'].head()
```

```python
df.groupby('contract')['Churn'].value_counts()
```

```python
for col in ['contract', 'payment_method', 'internet_service', 'tech_support']:
    df[col] = df[col].str.strip().str.lower().str.replace(r'[^\w\s]', '', regex=True)

df[['contract', 'payment_method', 'internet_service', 'tech_support']].head()
```

##**Recuento de evasiÃ³n por variables categÃ³ricas**

```python
def GraficarChurnPorCategoria(columna):
    nombresColumnas = {
        'customer_gender': 'GÃ©nero del Cliente',
        'customer_senior': 'Adulto Mayor',
        'contract': 'Tipo de Contrato',
        'payment_method': 'MÃ©todo de Pago',
        'internet_service': 'Servicio de Internet',
        'tech_support': 'Soporte TÃ©cnico'
    }

    titulo = f'EvasiÃ³n de Clientes segÃºn {nombresColumnas.get(columna, columna)}'

    plt.figure(figsize=(7, 5))
    sns.countplot(data=df, x=columna, hue='Churn', palette='Set2')
    plt.title(titulo, fontsize=15, fontweight='bold')
    plt.xlabel(nombresColumnas.get(columna, columna), fontsize=12)
    plt.ylabel('Cantidad', fontsize=12)
    plt.xticks(rotation=20)
    plt.legend(title='Churn', loc='upper right')
    sns.despine(top=True, right=True)
    plt.tight_layout()
    plt.show()
```

##**Conteo de evasiÃ³n por variables numÃ©ricas**

```python
def GraficaChurnVsVariableNumerica(variable):
    nombresColumnas = {
        'charges_monthly': 'Cargos Mensuales',
        'charges_total': 'Cargos Totales',
        'daily_accounts': 'Cuentas Diarias',
        'servicios_activos': 'Servicios Activos',
        'customer_senior': 'Adulto Mayor'
    }

    nombreBonito = nombresColumnas.get(variable, variable)
    titulo = f'{nombreBonito} vs EvasiÃ³n de Clientes'

    plt.figure(figsize=(7, 5))
    sns.boxplot(data=df, x='Churn', y=variable, palette='coolwarm')
    plt.title(titulo, fontsize=15, fontweight='bold')
    plt.xlabel('Churn (0 = No, 1 = SÃ­)', fontsize=12)
    plt.ylabel(nombreBonito, fontsize=12)
    sns.despine(top=True, right=True)
    plt.tight_layout()
    plt.show()
```

```python
GraficaChurnVsVariableNumerica('charges_total')
```

#ðŸ“„Informe final

##**1. IntroducciÃ³n**

Telecom X enfrenta una alta tasa de cancelaciÃ³n de clientes.  Este anÃ¡lisis busca identificar factores que a que los clientes tomen la decicion de cancelar los servicios de Telecom X, mediante la exploraciÃ³n de datos, visualizaciones y tratamiento previo.  
Los resultados ayudarÃ¡n a proponer estrategias de retenciÃ³n de los clientes.

##**2. Limpieza y tratamiento de datos**

El conjunto de datos original fue proporcionado en formato JSON e incluÃ­a informaciÃ³n anidada en forma de diccionarios, agrupando atributos en bloques como `customer`, `account`, `internet` y `phone`.

Para facilitar el anÃ¡lisis, se aplicÃ³ una transformaciÃ³n que extrajo las variables mÃ¡s relevantes de dichos diccionarios y se integraron al DataFrame principal como columnas planas.

Entre las acciones realizadas:

- Se eliminaron espacios innecesarios con `.str.strip()` y se estandarizÃ³ el texto en minÃºsculas con `.str.lower()` para unificar categorÃ­as.
- Se crearon nuevas columnas como:
  - `charges_monthly` y `charges_total` (facturaciÃ³n mensual y total).
  - `daily_accounts`, calculada dividiendo la facturaciÃ³n mensual entre 30 dÃ­as.
  - `customer_senior`, transformando el valor original en un campo numÃ©rico para adultos mayores.
- Se convirtieron valores binarios como `"yes"` y `"no"` en `1` y `0` para facilitar el anÃ¡lisis.

Este proceso asegurÃ³ que los datos estuvieran limpios, consistentes y listos para su exploraciÃ³n analÃ­tica.

##**3. AnÃ¡lisis Exploratorio de Datos**

Se realizÃ³ un anÃ¡lisis exploratorio para identificar patrones relevantes en los datos y comprender el perfil de los clientes que abandonan el servicio.

**EstadÃ­sticas Generales:**

- Las mÃ©tricas como media, mediana, y desviaciÃ³n estÃ¡ndar fueron calculadas para variables numÃ©ricas como `charges_total`, `charges_monthly` y `cuentas_diarias`.
- La mayorÃ­a de los clientes presentan cargos mensuales regulares, pero existen valores extremos que pueden corresponder a usuarios con paquetes mÃ¡s avanzados.

**DistribuciÃ³n de EvasiÃ³n:**

- Se visualizÃ³ la variable `Churn` mediante un grÃ¡fico de barras.
- Se identificÃ³ un porcentaje considerable de clientes que han cancelado el servicio.

**AnÃ¡lisis por CategorÃ­as:**

Se analizaron variables categÃ³ricas como `contract`, `payment_method`, `internet_service`, `tech_support`, `customer_gender` y `customer_senior`.

- Se observÃ³ una mayor tasa de evasiÃ³n en clientes con contratos mensuales, asÃ­ como en aquellos sin soporte tÃ©cnico.
- El gÃ©nero o la edad avanzada (`customer_senior`) no mostraron diferencias marcadas en la tasa de cancelaciÃ³n.

Este anÃ¡lisis permitiÃ³ delimitar los grupos con mayor propensiÃ³n a la evasiÃ³n y preparÃ³ el terreno para comparaciones con variables numÃ©ricas.


##**4. Conclusiones e Insights**

A partir del anÃ¡lisis realizado, se pueden destacar los siguientes hallazgos:

- **Contrato mensual**: Los clientes con este tipo de contrato tienen una probabilidad significativamente mayor de abandonar el servicio.
- **FacturaciÃ³n baja**: Los usuarios con cargos mensuales y diarios mÃ¡s bajos mostraron una mayor tendencia a cancelar.
- **Ausencia de soporte tÃ©cnico**: Aquellos clientes que no tienen acceso a soporte tÃ©cnico presentan tasas mÃ¡s altas de evasiÃ³n, lo cual destaca la importancia de este servicio.
- **MÃ©todo de pago electrÃ³nico**: Algunos mÃ©todos, como la facturaciÃ³n electrÃ³nica sin soporte humano, podrÃ­an correlacionarse con una experiencia de cliente menos satisfactoria.
